## İleri Toplulaştırma İşlemleri (Aggregate,filter,transform,apply)

```python
#aggregate

import pandas as pd
import numpy as np

a = pd.DataFrame({'gruplar':['A','B','C','A','B','C'],
                  'degisken1':[10,23,33,22,11,99],
                  'degisken2':[100,253,333,262,111,969],
                 }, columns = ['gruplar','degisken1','degisken2'])


a.groupby("gruplar")[["degisken1","degisken2"]].aggregate(["min",np.median,"max"])  # min ve max fonksiyonları pandasın içerisinde olduğu için tırnak içinde veya olduğu gibi bırakabiliriz.
                                                        # ancak pandasın içerisinde olmayan yani dışsal fonksiyon olduğu için np. aracılığıyla tanımlamamız lazım. (np.median gibi)
```

```python
import pandas as pd
a = pd.DataFrame({'gruplar':['A','B','C','A','B','C','D','D'],
                  'degisken1':[10,23,33,22,11,99,50,60],
                  'degisken2':[100,253,333,262,111,969,150,124],
                 }, columns = ['gruplar','degisken1','degisken2'])


a.groupby("gruplar").aggregate({"degisken1" : np.median , "degisken2" : np.mean })
```

```python
#filter
import pandas as pd
a = pd.DataFrame({'gruplar':['A','B','C','A','B','C','D','D'],
                  'degisken1':[10,23,33,22,11,99,50,60],
                  'degisken2':[100,253,333,262,111,969,150,124],
                 }, columns = ['gruplar','degisken1','degisken2'])

print(a,"\n")
print(a.groupby("gruplar").std(),"\n")  # grupların standart sapma değerleri

def filter_func(x):
    return x["degisken1"].std() > 9

a.groupby("gruplar").filter(filter_func)  #grupları tuttuk sonra bu grupları fonksiyona gönderdik , standart sapması 9 dan büyük olanlar filtrelendi.
```

```python
#transform : transform() işlevi, dönüştürülmüş değerlere sahip ve self ile aynı eksen uzunluğuna sahip bir Seriyi kendi kendine üreten işlevi çağırmak için kullanılır.
#kısacası transform() fonksiyonu ile sütun ve deişkenler bazında , bir fonksiyonda tanımladığımız işleme göre yapılan değişiklerdir. 

import pandas as pd
a = pd.DataFrame({'gruplar':['A','B','C','A','B','C','D','D'],
                  'degisken1':[10,23,33,22,11,99,50,60],
                  'degisken2':[100,253,333,262,111,969,150,124],
                 }, columns = ['gruplar','degisken1','degisken2'])

b = a.iloc[:,1:3]

print(b.mean())
print(b.std())
w
b.transform(lambda x: x-x.mean() / x.std())  #fonksiyon içerisinde sayısal işlemlerde bulunduğumuz için veri setinde kategorik değişkenin olmaması gerekir yoksa hata alırız. O yüzden gruplar değişkenini sildik
```
