```python
# DataFrame.drop(labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')


# index silme
import numpy as np
import pandas as pd

a = np.arange(5,30).reshape((5,5))
b = pd.DataFrame(a,columns = ["group-1","group-2","group-3","group-4","group-5"],index=[10,20,30,40,50])

b.columns = ("A","B","C","D","E")
b.index = ("a","b","c","d","e")

b.drop("b" , axis = 0 , inplace = True)  # inplace'in görevi : DataFrame'de yapılan işlemi kalıcı halde olmasını sağlar.Mesela bu işlemle b indeksinizi veri setinden kalıcı olarak sildik.
print(b)

l = ["c","e"]
b.drop(l , axis = 0 , inplace = True )  # inplace = False ise veri setini çıktıdan sonra tekrar çalıştırınca indeksin silinmeden önceki halini verir. True ise kalıcı olarak sildiği için veri setini ne zaman çağırırsak çağıralım değişken silindikten sonraki halini verir.
# drop metodu yardımıyla önce b indeksini kalıcı(inplace) olarak sildik.Sonra da istediğimiz 2 indeksi (c,e) fancy yöntemiyle kalıcı(inplace) olarak sildik.

print("\n", "is there column A: ","A" in b ,"\t", "is there column b : ", "F" in b ,"\n") # değişkenler(sütun) bazında sorgulama
b
```

```python
# bozuk ve kayıp verileri silme --> dropna() : Dropna kullanırken, NaN değerlerine (veya genel olarak null) sahip tüm girişleri kaldırır.

#DataFrame.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)

import pandas as pd
import numpy as np

arr = np.array([[10,20,np.nan,5],[3,np.nan,np.nan,7],[13,np.nan,4,11],[1.5,7.8,6,55]])
df = pd.DataFrame(arr, index = ['Index1','Index2','Index3','Index4'],columns = ['Column1','Column2','Column3','Column4'] )

print(df.dropna(),"\n")
print(df.dropna(axis=0),"\n")    #hangi satırda eksik değer varsa o satırı komple siler.

print(df.dropna(axis=1),"\n")    #hangi sütunda eksik değer varsa o sütunu komple siler.

print(df.dropna(how="all"),"\n")      # tüm gözlem birimi eksik değer olan yerleri siler.(df dataframe'de bir satırdaki tüm değerleri eksik değer alan satır olmadığından etkisi olmadı.)

print(df.dropna(subset = "Column3"),"\n")   # subset ile belirtilen sütundaki 'sadece' eksik değerleri siler tüm sütunu silmez ama belirtilen sütundaki eksik değerlerin olduğu indekslerdeki satırları da siler.
print(df.dropna(subset = ["Column3","Column4"]),"\n")
df
```

```python
# Bozuk yerleri nasıl doldururuz ?
print(df.fillna(value = 'no input'),"\n")
print(df.fillna(value = 999))
```

## Gözlem Ve Değişken Seçimi 
* <span style='color:blue'>**loc**</span>
* <span style='color:blue'>**iloc**</span>

```python
import numpy as np
import pandas as pd

a = np.random.randint(1,50, size = (10,7))
b = pd.DataFrame(a, columns = ["var1","var2","var3","var4","var5","var6","var7"])
b
```

```python
# loc: Pandas dataframe'i üzerinde hem gözlem hem değişken seçimi için index bağımsız seçim yapmak üzere loc kullanılır.
# iloc label (etiket,kolon,row vs ismiyle) bagimli çalışmaktadır.
#normalde indeks işlemi yaparken [x:y] x den başlanır ve y'ye kadar(y dahil değil) veri çağrılır.Ancak loc komutunda y'yi de dahil ediyoruz.


import numpy as np
import pandas as pd

a = np.random.randint(1,30, size = (10,4))
b = pd.DataFrame(a, columns = ["var1","var2","var3","var4"])

print(b.loc[0:3],"\n") 
                   

    
# !!! eğer değişkenlere göre bir indeksleme yapacaksak loc'u kullanmalıyız.Aksi taktirde iloc komutunda hataya maruz kalırız.   
print(b.loc[0:4,"var2"])    
    
    
# iloc: Pandas dataframe'inde index'lere bağlı kalarak gözlem ve değişken seçimi yapmak için iloc kullanılır.

print("\n",b.iloc[0:3])


# iloc ile elemanlara erişme

print("\n",b.iloc[0,0],",",b.iloc[0,3],"\n\n",b.iloc[:3,:2])
print("\n",b,"\n")
b.iloc[2,0] = 15
print("\n",b,"\n")
b.iloc[:10,:4]
```
